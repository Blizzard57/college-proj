```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 10000
val = 1
delta = 0.001

def potential(k,x):
    t = 0.5 * k * x**2
    return t

def force(k,x):
    t = -1 * k * x
    return t

p = np.array([4])
x = np.array([0])
t = np.array([0])
e = np.array([8])
k = np.array([8])
u = np.array([0])
for i in range(n):
    temp2 = x[-1] + delta*p[-1]
    x = np.append(x,temp2)
    temp = p[-1] + delta*force(val,x[-1])
    p = np.append(p,temp)
    temp3 = t[-1] + delta
    t = np.append(t,temp3)
    temp4 = 0.5*(x[-1]**2) + (p[-1]**2)*0.5
    e = np.append(e,temp4)
    temp5 = 0.5*(x[-1]**2)
    u = np.append(u,temp5)
    temp6 = (p[-1]**2)*0.5
    k = np.append(k,temp6)
plt.plot(t,e)
plt.plot(t,u)
plt.plot(t,k)
plt.show()
plt.plot(t,x)
plt.plot(t,p)
plt.show()
```
